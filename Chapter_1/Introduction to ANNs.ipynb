*An artificial Neural Network is a computational model designed to simulate how human brains processes information.*


*Neurons are the fundamental units that process information in an artifical neural network.*

As in the image above, I have used light bulbs to represent neurons. I think of neurons are light bulbs which glow with information. As the neurons in ANN processes data, some neurons light up intensely, signaling strong connections and robust information processing, while others glow faintly, indicating weaker interactions. This metaphor illuminates the dynamic interplay within the artificial neural network, where neurons influence each other’s activity, creating intricate patterns of light.


*Weights are parameters that scale the input signals to a neuron, determining the neuron’s output and thereby influencing the artificial neural network’s learning and decision-making processes.*

Think of weights like dimmers that adjust how bright the light bulb should shine (similar to how a resistor manages electrical flow in a circuit). Turning the dimmer up allows more electricity to flow, making the bulb brighter (stronger influence). Turning it down reduces the flow, making the bulb dimmer (weaker influence).  During training, these weights are adjusted (updated) to make certain connections stronger and others weaker, similar to how the brain emphasizes essential pathways through learning.

*Bias in neural networks is an additional constant added to the weighted sum of inputs that adjusts the threshold at which a neuron activates.*

Think of bias as a battery that provides the baseline voltage. When a neuron receives input signals, these are combined and adjusted by their weights, similar to adjusting the brightness with a dimmer switch. However, if this combined signal is still below what’s needed for the neuron to fire (turn on), the bias acts like a minimum voltage guarantee, ensuring the neuron has enough input to activate. This is crucial for the neuron to function correctly, especially in cases where input signals are weak or insufficient on their own.

*Activation functions in neural networks compute a neuron’s output by applying a non-linear mathematical transformation to the sum of the weighted inputs and the bias. This transformation enables outputs that do not have a direct, proportional relationship with inputs, allowing the network to handle complex patterns and interactions within data.*

Think of activation function as a smart switch that ultimately decides whether the bulb lights up, based on the total incoming power adjusted for by the dimmers and the voltage. This smart switch has sophisticated features:
* **For a Sigmoid Activation**:` It behaves like a sensitive dimmer that gradually increases the bulb’s brightness as power approaches a critical level and then caps the brightness at a maximum limit. This provides a smooth transition from off to fully lit, mimicking the sigmoid’s smooth curve.
* **For ReLU Activation**: It acts like a threshold switch that keeps the bulb off for any negative current but turns it on and adjusts brightness linearly with positive currents. This switch directly translates increased input into brighter light without any cap, reflecting ReLU’s linear characteristics once the threshold is exceeded.

The term “smart switch” effectively captures the dynamic and responsive nature of activation functions. These switches aren’t just binary (not only ON/OFF); they adjust their responses based on the inputs’ strength and quality, enabling the neural network to handle complex problems.

## Code Snippets:

```python

import matplotlib.pyplot as plt
import numpy as np

# Define the ReLU function
def ReLU(x):
    return np.maximum(0, x)

# Generate a range of 400 values between -10 and 10
x = np.linspace(-10, 10, 400)
y = ReLU(x)

# Plot the ReLU function
plt.figure(figsize=(8, 4)) # Set the size of the plot
plt.plot(x, y, label="ReLU Function") # Plot the ReLU function
plt.xlabel("Input") # Set the x-axis label
plt.ylabel("Output") # Set the y-axis label
plt.title("ReLU Activation Function") # Set the title of the plot
plt.legend() # Show the legend
plt.show() # Display the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the Sigmoid function
def Sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate a range of 400 values between -10 and 10
x = np.linspace(-10, 10, 400)
y = Sigmoid(x)

# Plot the Sigmoid function
plt.figure(figsize=(8, 4)) # Set the size of the plot
plt.plot(x, y, label="Sigmoid Function") # Plot the Sigmoid function
plt.xlabel("Input") # Set the x-axis label
plt.ylabel("Output") # Set the y-axis label
plt.title("Sigmoid Activation Function") # Set the title of the plot
plt.legend() # Show the legend
plt.show() # Display the plot
```
