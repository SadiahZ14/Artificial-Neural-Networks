*An artificial Neural Network is a computational model designed to simulate how human brains processes information.*


*Neurons are the fundamental units that process information in an artifical neural network.*

As in the image above, I have used light bulbs to represent neurons. I think of neurons are light bulbs which glow with information. As the neurons in ANN processes data, some neurons light up intensely, signaling strong connections and robust information processing, while others glow faintly, indicating weaker interactions. This metaphor illuminates the dynamic interplay within the artificial neural network, where neurons influence each other’s activity, creating intricate patterns of light.


*Weights are parameters that scale the input signals to a neuron, determining the neuron’s output and thereby influencing the artificial neural network’s learning and decision-making processes.*

Think of weights like dimmers that adjust how bright the light bulb should shine (similar to how a resistor manages electrical flow in a circuit). Turning the dimmer up allows more electricity to flow, making the bulb brighter (stronger influence). Turning it down reduces the flow, making the bulb dimmer (weaker influence).  During training, these weights are adjusted (updated) to make certain connections stronger and others weaker, similar to how the brain emphasizes essential pathways through learning.

*Bias in neural networks is an additional constant added to the weighted sum of inputs that adjusts the threshold at which a neuron activates.*

Think of bias as the baseline voltage. When a neuron receives input signals, these are combined and adjusted by their weights, similar to adjusting the brightness with a dimmer switch. However, if this combined signal is still below what’s needed for the neuron to fire (turn on), the bias acts like a minimum voltage guarantee, ensuring the neuron has enough input to activate. This is crucial for the neuron to function correctly, especially in cases where input signals are weak or insufficient on their own.

4.	Activation Functions as Sophisticated Dimmer Switches: The activation function acts like a high-tech dimmer switch that controls when the bulb lights up, based on the combined current from the inputs and the bias. This switch does more than just toggle the bulb on or off:
	•	Sigmoid Activation: This function is like a dimmer that increases the bulb’s brightness gradually as the input nears a critical threshold, then saturates at maximum brightness. It ensures the output (brightness) smoothly transitions from off to fully lit, which helps capture subtle differences in input strength.
	•	ReLU Activation: This function keeps the bulb off for all negative currents but allows it to brighten proportionally to the current once it becomes positive. This means the bulb’s brightness directly reflects the positive input strength without any upper limit, providing a clear and linear response once the threshold is surpassed.

Indeed, using the term “smart switch” can still be very effective in this analogy—it emphasizes the advanced capabilities and nuanced control provided by activation functions in a neural network. Let’s reintegrate that term to enhance the analogy further:

	1.	Neurons as Light Bulbs: Think of each neuron as a light bulb that illuminates when it receives enough power. The illumination represents the neuron’s activation, crucial for transmitting information within the network.
	2.	Weights as Dimmers: Attached to each neuron are multiple inputs, each regulated by its own dimmer switch (the weight). These dimmers control how much each input affects the neuron, much like adjusting how much current flows through each wire.
	3.	Bias as Baseline Voltage: Bias acts like a battery or a constant voltage source connected to the light bulb. It ensures there’s always some minimal energy available, making it easier for the neuron to reach the activation threshold with the combined inputs.
	4.	Activation Functions as Smart Switches: Here, the activation function is a smart switch that ultimately decides whether the bulb lights up, based on the total incoming power adjusted for by the dimmers and the battery. This smart switch has sophisticated features:
	•	For a Sigmoid Activation: It behaves like a sensitive dimmer that gradually increases the bulb’s brightness as power approaches a critical level and then caps the brightness at a maximum limit. This provides a smooth transition from off to fully lit, mimicking the sigmoid’s smooth curve.
	•	For ReLU Activation: It acts like a threshold switch that keeps the bulb off for any negative current but turns it on and adjusts brightness linearly with positive currents. This switch directly translates increased input into brighter light without any cap, reflecting ReLU’s linear characteristics once the threshold is exceeded.

The term “smart switch” effectively captures the dynamic and responsive nature of activation functions. These switches aren’t just binary; they adjust their responses based on the inputs’ strength and quality, enabling the neural network to handle complex, non-linear problems more adeptly. This advanced functionality allows the network to perform sophisticated data processing tasks that mimic some aspects of human cognitive functions.
